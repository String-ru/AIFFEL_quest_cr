```python
from skimage.io import imread
from skimage.color import rgb2gray

image = imread('universe.jpg')  # 파일 이름을 확인하고 수정하세요.
gray_image = rgb2gray(image)
```

```python
imshow(image)
```

```python
imshow(gray_image)
```

```python
# Edge Detection
edge = np.array([[-1, -1, -1],
                    [-1, 8, -1],
                    [-1, -1, -1]])
# Sharpen
sharpen = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])
# Gaussian Blur
blur = (1 / 16.0) * np.array([[1., 2., 1.],
                                 [2., 4., 2.],
                                 [1., 2., 1.]])
```

```python
# Edge Detection 적용
edge_im = convolve2d(gray_image, edge, 'valid')
imshow(edge_im)
```

```python
sharp_im = convolve2d(gray_image, sharpen, 'valid')
imshow(sharp_im)
```

```python
# Gaussian Blur filter를 이미지에 적용해 보세요.
blur_im = convolve2d(gray_image, blur, 'valid')
imshow(blur_im)
```
