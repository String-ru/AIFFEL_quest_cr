# Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨ Î∞è data Î≥ëÌï©

```python
# 1Ô∏è‚É£ Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨
for df in [disney_df, amazon_df, netflix_df]:
    df = df.copy()  # Í≤ΩÍ≥† Î∞©ÏßÄÎ•º ÏúÑÌï¥ Î≥µÏÇ¨Î≥∏ ÏÇ¨Ïö©

    # 'date_added' Ïª¨ÎüºÏù¥ Ï°¥Ïû¨ÌïòÎ©¥ ÏÇ≠Ï†ú
    if 'date_added' in df.columns:
        df = df.drop(columns=['date_added'])

    # 'director'ÏôÄ 'cast'Ïùò Í≤∞Ï∏°ÏπòÎ•º 'Unknown'ÏúºÎ°ú Ï±ÑÏö∞Í∏∞
    df['director'] = df['director'].fillna('Unknown')
    df['cast'] = df['cast'].fillna('Unknown')

    # 'duration'Ïùò Í≤∞Ï∏°ÏπòÍ∞Ä 3Í∞úÎøêÏù¥ÎØÄÎ°ú Ìï¥Îãπ Ìñâ Ï†úÍ±∞
    df = df.dropna(subset=['duration'])

# 2Ô∏è‚É£ Îç∞Ïù¥ÌÑ∞ Î≥ëÌï©
combined_df = pd.concat([disney_df, amazon_df, netflix_df], ignore_index=True)

# ÌôïÏù∏
print("‚úÖ ÏµúÏ¢Ö Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞:", combined_df.shape)
print(combined_df.head())
```

```python
# Ï§ëÎ≥µÎêú ÏòÅÌôî/TV ÌîÑÎ°úÍ∑∏Îû®ÏùÑ ÌïòÎÇòÎßå ÎÇ®Í∏∞Í≥† ÏÇ≠Ï†ú
# Ïö∞ÏÑ†ÏàúÏúÑ ÏÑ§Ï†ï: Netflix > Amazon Prime > Disney+
platform_order = ['Netflix', 'Amazon', 'Disney']

# Ï§ëÎ≥µÎêú titleÏóê ÎåÄÌï¥ platform Í∏∞Ï§ÄÏúºÎ°ú Ïö∞ÏÑ†ÏàúÏúÑÎ•º Ï†ÅÏö©ÌïòÏó¨ ÎÇ®Í∏∏ ÌñâÏùÑ Í≤∞Ï†ï
df_sorted = df.sort_values(by=['title', 'platform'], ascending=[True, True])
df_sorted['platform'] = pd.Categorical(df_sorted['platform'], categories=platform_order, ordered=True)

# Ï§ëÎ≥µÎêú titleÏùÑ Í∞ÄÏßÑ ÌñâÏóêÏÑú platform ÏàúÏÑúÎåÄÎ°ú Ï§ëÎ≥µ Ï†úÍ±∞
df_deduplicated = df_sorted.drop_duplicates(subset=['title'], keep='first')

# Í≤∞Í≥º ÌôïÏù∏
print(f'Ï§ëÎ≥µ Ï†úÍ±∞ ÌõÑ Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞: {df_deduplicated.shape}')
```

# Ìïú ÏòÅÌôî/TVÌîÑÎ°úÍ∑∏Îû® Îãπ Ìïú Ïû•Î•¥ Î∞∞Ï†ïÌïòÍ∏∞

listed_in colunmÏô∏ÏóêÎèÑ

categoryÏôÄ category_list	columnÍ≥º top_category column ÌôúÏö©

```python
# 4Ô∏è‚É£ ÏÑ∏Î∂Ä Ïû•Î•¥Î•º ÎåÄÎ∂ÑÎ•ò Ïπ¥ÌÖåÍ≥†Î¶¨Î°ú Î≥ÄÌôòÌïòÎäî Ìï®Ïàò
def map_genres_to_categories(listed_in, genre_mapping):
    # ÏûëÏùÄÎî∞Ïò¥Ìëú Ï†úÍ±∞
    listed_in = listed_in.replace("'", "")

    # Í≤∞Í≥ºÎ°ú ÎåÄÎ∂ÑÎ•ò Ïπ¥ÌÖåÍ≥†Î¶¨ Ï†úÎ™©Îßå Îã¥ÏùÑ Î¶¨Ïä§Ìä∏
    categories = []

    # ÏÑ∏Î∂Ä Ïû•Î•¥Îì§ÏùÑ ÎåÄÎ∂ÑÎ•ò Ïπ¥ÌÖåÍ≥†Î¶¨Î°ú Îß§Ìïë
    for category, genres in genre_mapping.items():
        for genre in genres:
            if genre in listed_in:
                categories.append(category)
                break  # ÌïòÎÇòÏùò ÎåÄÎ∂ÑÎ•ò Ïπ¥ÌÖåÍ≥†Î¶¨Îßå Ï∂îÍ∞Ä (Ï§ëÎ≥µ Î∞©ÏßÄ)

    # ÎåÄÎ∂ÑÎ•ò Ïπ¥ÌÖåÍ≥†Î¶¨Îì§Ïù¥ ÏóÜÏúºÎ©¥ Îπà Î¨∏ÏûêÏó¥ Î∞òÌôò
    return ', '.join(categories) if categories else ''

# 5Ô∏è‚É£ 'listed_in' Ïª¨ÎüºÏùÑ ÎåÄÎ∂ÑÎ•ò Ïπ¥ÌÖåÍ≥†Î¶¨Î°ú Î≥ÄÌôò
df_deduplicated['category'] = df_deduplicated['listed_in'].apply(lambda x: map_genres_to_categories(x, genre_mapping))

# 6Ô∏è‚É£ Í≤∞Í≥º ÌôïÏù∏
print(df_deduplicated[['title', 'platform', 'listed_in', 'category']])

from IPython.display import display

display(df_deduplicated.head(20))
```

```python
import pandas as pd
from collections import Counter

# df_deduplicatedÎ•º Í∏∞Î∞òÏúºÎ°ú ÏûëÏóÖ
# Í∞Å ÏûëÌíàÏùò categoryÎ•º ÏâºÌëúÎ°ú Î∂ÑÌï†ÌïòÏó¨ Î¶¨Ïä§Ìä∏Î°ú Î≥ÄÌôò
df_deduplicated["category_list"] = df_deduplicated["listed_in"].apply(lambda x: x.split(", "))

# Î™®Îì† categoryÎ•º Î™®ÏïÑÏÑú ÎπàÎèÑÏàò Í≥ÑÏÇ∞
all_categories = [cat for sublist in df_deduplicated["category_list"] for cat in sublist]
category_counts = Counter(all_categories)

# Ïπ¥ÌÖåÍ≥†Î¶¨Î•º ÎπàÎèÑÏàò Í∏∞Ï§ÄÏúºÎ°ú Ï†ïÎ†¨
top_categories = sorted(category_counts.items(), key=lambda x: x[1], reverse=True)
print("Ïπ¥ÌÖåÍ≥†Î¶¨ ÎπàÎèÑÏàò ÏàúÏúÑ:", top_categories)

# Í∞Å ÏûëÌíàÏóê ÎåÄÌï¥ Í∞ÄÏû• Ïö∞ÏÑ†ÏàúÏúÑÍ∞Ä ÎÜíÏùÄ ÎåÄÎ∂ÑÎ•ò Î∞∞Ï†ï
def get_top_category(categories):
    sorted_categories = sorted(categories, key=lambda x: category_counts[x], reverse=True)
    return sorted_categories[0] if sorted_categories else None

df_deduplicated["top_category"] = df_deduplicated["category_list"].apply(get_top_category)

# Í≤∞Í≥º Ï∂úÎ†•
print(df_deduplicated[["title", "top_category"]])

display(df_deduplicated.head(20))
```

#cast column Î∂ÑÎ¶¨

- 'cast' Ïª¨ÎüºÏùò Í∞íÏùÑ ','Î°ú Î∂ÑÎ¶¨ÌïòÏó¨ Î∞∞Ïö∞ Ïù¥Î¶ÑÏùÑ Î¶¨Ïä§Ìä∏ ÌòïÌÉúÎ°ú 'actor' Ïª¨ÎüºÏóê Ï†ÄÏû• (applyÏôÄ lambda Ìï®Ïàò)

- 'actor' Ïª¨ÎüºÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Í∞Å Î∞∞Ïö∞Í∞Ä Î≥ÑÎèÑÏùò ÌñâÏùÑ Í∞ñÎèÑÎ°ù ÏßÑÌñâ

- explodeÎ•º ÏÇ¨Ïö©ÌïòÏó¨ 'actor' Ïª¨ÎüºÏùò Î¶¨Ïä§Ìä∏ Í∞íÏùÑ Ïó¨Îü¨ ÌñâÏúºÎ°ú ÌôïÏû•ÌñàÏäµÎãàÎã§.

- 'actor' Ïª¨ÎüºÏùò Í∞íÏóêÏÑú str.strip() ÏÇ¨Ïö©ÌïòÏó¨ Î∞∞Ïö∞ Ïù¥Î¶ÑÏùò ÏïûÎí§ Í≥µÎ∞±ÏùÑ Ï†úÍ±∞

```python
# 'cast' Ïª¨ÎüºÏùÑ ',' Í∏∞Ï§ÄÏúºÎ°ú Î∂ÑÎ¶¨Ìï¥ÏÑú Î¶¨Ïä§Ìä∏Î°ú ÎßåÎì§Í∏∞
df['actor'] = df['cast'].apply(lambda x: str(x).split(','))

# 'actor' Ïª¨ÎüºÏóêÏÑú ÎåÄÍ¥ÑÌò∏ Ï†úÍ±∞
df['actor'] = df['actor'].apply(lambda x: [item.replace('[', '').replace(']', '').strip() for item in x])

# Í∞ôÏùÄ ÏòÅÌôîÏóê Ï∂úÏó∞Ìïú Î∞∞Ïö∞(cast Ïª¨ÎüºÏùò Î¶¨Ïä§Ìä∏)Î•º Ïó¨Îü¨ ÌñâÏúºÎ°ú ÌôïÏû•
df_exploded = df.explode('actor')

# Î∞∞Ïö∞ Ïù¥Î¶Ñ ÏïûÎí§ Í≥µÎ∞± Ï†úÍ±∞
df_exploded['actor'] = df_exploded['actor'].str.strip()

# Í≤∞Í≥º ÌôïÏù∏ (ÌÉÄÏù¥ÌãÄÍ≥º Î∞∞Ïö∞ Ïª¨ÎüºÎßå Î≥¥Í∏∞)
df_exploded[['title', 'actor']].head(50)
```

#listed_in column Î∂ÑÎ¶¨

- 'listed_in' Ïª¨ÎüºÏùò Í∞íÏùÑ ÏâºÌëú(,)Î°ú Î∂ÑÎ¶¨ÌïòÏó¨ Ïû•Î•¥ Î¶¨Ïä§Ìä∏Î•º ÏÉùÏÑ±ÌïòÍ≥†, 'genre_list' Ïª¨ÎüºÏóê Ï†ÄÏû•

- str.split()Í≥º explode()Î•º Ïù¥Ïö©ÌïòÏó¨ Ïû•Î•¥ Ï†ïÎ≥¥Î•º Î∂ÑÎ¶¨ÌïòÍ≥† Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ÌôïÏû•

- 'title' Ïª¨ÎüºÏùÑ Ïú†ÏßÄÌïòÎ©¥ÏÑú 'genre_list' Ïª¨ÎüºÏùÑ Ïó¨Îü¨ ÌñâÏúºÎ°ú ÌôïÏû•ÌïòÏó¨ ÏûëÌíàÍ≥º Ïû•Î•¥ Ï†ïÎ≥¥Î•º Ïó∞Í≤∞

- explode()Î•º ÏÇ¨Ïö©ÌïòÏó¨ Í∞Å Ïû•Î•¥Í∞Ä Î≥ÑÎèÑÏùò ÌñâÏùÑ Í∞ñÎèÑÎ°ù Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Î≥ÄÌòï

```python
# 'listed_in' Ïª¨ÎüºÏùò Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏùÑ Î¨∏ÏûêÏó¥Î°ú Î≥ÄÌôòÌïòÍ≥† NaN Í∞íÏùÑ Îπà Î¨∏ÏûêÏó¥Î°ú ÎåÄÏ≤¥
df['listed_in'] = df['listed_in'].astype(str).fillna('')

# 'listed_in' Ïª¨ÎüºÏùò Í∞íÏùÑ ÏâºÌëú(`, `)Î°ú Î∂ÑÎ¶¨ÌïòÏó¨ Ïû•Î•¥ Î¶¨Ïä§Ìä∏ ÏÉùÏÑ±
df['genre_list'] = df['listed_in'].apply(lambda x: x.split(', ')) # explode Ïù¥Ï†ÑÏóê str.split Ï†ÅÏö©

# titleÏùÑ Ïú†ÏßÄÌïòÎ©¥ÏÑú genre_listÎ•º Ïó¨Îü¨ ÌñâÏúºÎ°ú ÌôïÏû•
df_exploded_genre = df[['title', 'genre_list']].explode('genre_list')

df_exploded_genre['genre_list'] = df_exploded_genre['genre_list'].str.replace(r"[\[\]]", "", regex=True)

# Í≤∞Í≥º ÌôïÏù∏
df_exploded_genre.head(10)
```
