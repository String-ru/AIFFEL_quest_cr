# 한 영화/TV프로그램 당 한 장르 배정하기

listed_in colunm외에도

category와 category_list	column과 top_category column 활용

```python
# 4️⃣ 세부 장르를 대분류 카테고리로 변환하는 함수
def map_genres_to_categories(listed_in, genre_mapping):
    # 작은따옴표 제거
    listed_in = listed_in.replace("'", "")

    # 결과로 대분류 카테고리 제목만 담을 리스트
    categories = []

    # 세부 장르들을 대분류 카테고리로 매핑
    for category, genres in genre_mapping.items():
        for genre in genres:
            if genre in listed_in:
                categories.append(category)
                break  # 하나의 대분류 카테고리만 추가 (중복 방지)

    # 대분류 카테고리들이 없으면 빈 문자열 반환
    return ', '.join(categories) if categories else ''

# 5️⃣ 'listed_in' 컬럼을 대분류 카테고리로 변환
df_deduplicated['category'] = df_deduplicated['listed_in'].apply(lambda x: map_genres_to_categories(x, genre_mapping))

# 6️⃣ 결과 확인
print(df_deduplicated[['title', 'platform', 'listed_in', 'category']])

from IPython.display import display

display(df_deduplicated.head(20))
```

```python
import pandas as pd
from collections import Counter

# df_deduplicated를 기반으로 작업
# 각 작품의 category를 쉼표로 분할하여 리스트로 변환
df_deduplicated["category_list"] = df_deduplicated["listed_in"].apply(lambda x: x.split(", "))

# 모든 category를 모아서 빈도수 계산
all_categories = [cat for sublist in df_deduplicated["category_list"] for cat in sublist]
category_counts = Counter(all_categories)

# 카테고리를 빈도수 기준으로 정렬
top_categories = sorted(category_counts.items(), key=lambda x: x[1], reverse=True)
print("카테고리 빈도수 순위:", top_categories)

# 각 작품에 대해 가장 우선순위가 높은 대분류 배정
def get_top_category(categories):
    sorted_categories = sorted(categories, key=lambda x: category_counts[x], reverse=True)
    return sorted_categories[0] if sorted_categories else None

df_deduplicated["top_category"] = df_deduplicated["category_list"].apply(get_top_category)

# 결과 출력
print(df_deduplicated[["title", "top_category"]])

display(df_deduplicated.head(20))
```

#cast column 분리

- 'cast' 컬럼의 값을 ','로 분리하여 배우 이름을 리스트 형태로 'actor' 컬럼에 저장 (apply와 lambda 함수)

- 'actor' 컬럼을 기준으로 각 배우가 별도의 행을 갖도록 진행

- explode를 사용하여 'actor' 컬럼의 리스트 값을 여러 행으로 확장했습니다.

- 'actor' 컬럼의 값에서 str.strip() 사용하여 배우 이름의 앞뒤 공백을 제거

```python
# 'cast' 컬럼을 ',' 기준으로 분리해서 리스트로 만들기
df['actor'] = df['cast'].apply(lambda x: str(x).split(','))

# 'actor' 컬럼에서 대괄호 제거
df['actor'] = df['actor'].apply(lambda x: [item.replace('[', '').replace(']', '').strip() for item in x])

# 같은 영화에 출연한 배우(cast 컬럼의 리스트)를 여러 행으로 확장
df_exploded = df.explode('actor')

# 배우 이름 앞뒤 공백 제거
df_exploded['actor'] = df_exploded['actor'].str.strip()

# 결과 확인 (타이틀과 배우 컬럼만 보기)
df_exploded[['title', 'actor']].head(50)
```

- str.split()과 explode()를 이용하여 장르 정보를 분리하고 데이터프레임을 확장
