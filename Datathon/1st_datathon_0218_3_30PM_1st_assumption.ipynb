```python
import pandas as pd
import re
import numpy as np

# 3Í∞ú Îç∞Ïù¥ÌÑ∞ÏÖã Ìï©ÏπòÍ∏∞
df = pd.concat([netflix_df, amazon_df, disney_df], ignore_index=True)

# ÌïÑÏöîÌïú Ïª¨Îüº ÏÑ†ÌÉù
df = df[['title', 'platform', 'listed_in', 'cast']]

# Ï§ëÎ≥µÍ∞í Ï†úÍ±∞ (drop_duplicates Ïã§Ìñâ Ï†ÑÏóê Ïª¨ÎüºÎì§ÏùÑ Î¨∏ÏûêÏó¥Î°ú Î≥ÄÌôò)
df['title'] = df['title'].astype(str)
df['platform'] = df['platform'].astype(str)
df['listed_in'] = df['listed_in'].astype(str)
df['cast'] = df['cast'].astype(str)
df.drop_duplicates(inplace=True)

# title, listed_in, platform Ïª¨ÎüºÏùÑ Î¶¨Ïä§Ìä∏ ÌòïÏãùÏúºÎ°ú Î≥ÄÌôò
df["title"] = df["title"].fillna("").apply(lambda x: x.split(", ") if x else [])
df["listed_in"] = df["listed_in"].fillna("").apply(lambda x: x.split(", ") if x else [])
df["platform"] = df["platform"].fillna("").apply(lambda x: x.split(", ") if x else [])

# cast Ïª¨ÎüºÏùÄ Í∏∞Ï°¥ Î¶¨Ïä§Ìä∏ ÌòïÌÉú Ïú†ÏßÄ (Î¨∏ÏûêÏó¥Î°ú Ï†ÄÏû•ÎêòÏóàÏùÑ Í≤ΩÏö∞ Î≥ÄÌôò)
# cast Ïª¨ÎüºÏùò Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏùÑ Î¨∏ÏûêÏó¥Î°ú Î≥ÄÌôò
df['cast'] = df['cast'].astype(str)
# 'cast' Ïª¨ÎüºÏóêÏÑú NaN Í∞íÏùÑ 'unknown'ÏúºÎ°ú Î≥ÄÌôò
df['cast'] = df['cast'].apply(lambda x: ['unknown'] if x == 'nan' else eval(x) if x.startswith("[") else x.split(", "))


# ÏïåÌååÎ≤≥(A-Z, a-z) Î∞è Ïà´Ïûê(0-9)Îßå ÎÇ®Í∏∞Í≥† ÎÇòÎ®∏ÏßÄ Î¨∏Ïûê Ï†úÍ±∞
def clean_text_list(text_list):
    return ["".join(re.findall(r"[A-Za-z0-9\s]", text)).strip() for text in text_list if text]

df["title"] = df["title"].apply(clean_text_list)
df["listed_in"] = df["listed_in"].apply(clean_text_list)
df["platform"] = df["platform"].apply(clean_text_list)
df["cast"] = df["cast"].apply(clean_text_list)

# title, listed_in, platform Ïª¨ÎüºÏóêÏÑú ÎπÑÏñ¥ÏûàÎäî Î¶¨Ïä§Ìä∏([])Í∞Ä ÏûàÎäî Ìñâ ÏÇ≠Ï†ú
# cast Ïª¨ÎüºÏùÄ Ïù¥ÎØ∏ null Í∞íÏù¥ 'unknown'ÏúºÎ°ú Ï±ÑÏõåÏ†∏ ÏûàÏúºÎØÄÎ°ú Ï†úÏô∏
df_cleaned = df[(df["title"].apply(len) > 0) & (df["listed_in"].apply(len) > 0) & (df["platform"].apply(len) > 0)]

# Í≤∞Í≥º ÌôïÏù∏
df_cleaned.head()
```

```python
import pandas as pd

# ÏòàÏãú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ
data = {
    'title': ['Movie A', 'Movie B'],
    'cast': [['Actor 1', 'Actor 2'], ['Actor 3', 'Actor 4']],
    'listed_in': [['Action', 'Comedy'], ['Drama', 'Horror']]
}
df = pd.DataFrame(data)

# 1. 'cast' Ïª¨ÎüºÏùÑ Í∞úÎ≥Ñ Î∞∞Ïö∞Î°ú Ï™ºÍ∞úÍ≥†, 'listed_in'ÏùÑ 58Í∞ú Ïπ¥ÌÖåÍ≥†Î¶¨Ïóê Îß§Ìïë
expanded_rows = []

for _, row in df.iterrows():
    title = row['title']
    actors = row['cast']
    genres = row['listed_in']

    # Í∞Å Î∞∞Ïö∞ÏôÄ ÏòÅÌôîÏóê ÎåÄÌï¥, Ïû•Î•¥Î•º Î∞òÎ≥µÌïòÏó¨ ÏÉàÎ°úÏö¥ ÌñâÏùÑ ÏÉùÏÑ±
    for actor in actors:
        for genre in genres:
            # Ïû•Î•¥Î•º Ïπ¥ÌÖåÍ≥†Î¶¨Ïóê ÎßûÍ≤å Î≥ÄÍ≤Ω
            for category, subgenres in genre_mapping.items():
                if genre in subgenres:
                    expanded_rows.append({'title': title, 'actor': actor, 'genre_category': category})

# ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
expanded_df = pd.DataFrame(expanded_rows)

# Í≤∞Í≥º ÌôïÏù∏
print(expanded_df)
```
