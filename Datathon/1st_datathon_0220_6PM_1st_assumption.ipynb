# 5. ê²°ì¸¡ì¹˜ ì²˜ë¦¬ ë° data ë³‘í•©

ê²°ì¸¡ì¹˜ ì²˜ë¦¬ ë° ë°ì´í„° ë³‘í•© ì™„ë£Œ, ìµœì¢… ë°ì´í„° 19,925í–‰ 10ì»¬ëŸ¼ ìƒì„±. âœ…

```python
# 1ï¸âƒ£ ê²°ì¸¡ì¹˜ ì²˜ë¦¬
for df in [disney_df, amazon_df, netflix_df]:
    df = df.copy()  # ê²½ê³  ë°©ì§€ë¥¼ ìœ„í•´ ë³µì‚¬ë³¸ ì‚¬ìš©

    # 'date_added' ì»¬ëŸ¼ì´ ì¡´ì¬í•˜ë©´ ì‚­ì œ
    if 'date_added' in df.columns:
        df = df.drop(columns=['date_added'])

    # 'director'ì™€ 'cast'ì˜ ê²°ì¸¡ì¹˜ë¥¼ 'Unknown'ìœ¼ë¡œ ì±„ìš°ê¸°
    df['director'] = df['director'].fillna('Unknown')
    df['cast'] = df['cast'].fillna('Unknown')

    # 'duration'ì˜ ê²°ì¸¡ì¹˜ê°€ 3ê°œë¿ì´ë¯€ë¡œ í•´ë‹¹ í–‰ ì œê±°
    df = df.dropna(subset=['duration'])

# 2ï¸âƒ£ ë°ì´í„° ë³‘í•©
combined_df = pd.concat([disney_df, amazon_df, netflix_df], ignore_index=True)

# í™•ì¸
print("âœ… ìµœì¢… ë°ì´í„° í¬ê¸°:", combined_df.shape)
print(combined_df.head())
```

# 6. ì¤‘ë³µë˜ëŠ” ì˜í™”/TVí”„ë¡œê·¸ë¨ ì œê±°

ì¤‘ë³µëœ ì˜í™”/TV í”„ë¡œê·¸ë¨ì„ í”Œë«í¼ ìš°ì„ ìˆœìœ„ë¡œ ì •ë¦¬í•˜ì—¬ ì¤‘ë³µ ì œê±° ì™„ë£Œ, ìµœìµœì¢… ë°ì´í„° 8,804í–‰ 9ì»¬ëŸ¼. âœ…

```python
# ì¤‘ë³µëœ ì˜í™”/TV í”„ë¡œê·¸ë¨ì„ í•˜ë‚˜ë§Œ ë‚¨ê¸°ê³  ì‚­ì œ
# ìš°ì„ ìˆœìœ„ ì„¤ì •: Netflix > Amazon Prime > Disney+
platform_order = ['Netflix', 'Amazon', 'Disney']

# ì¤‘ë³µëœ titleì— ëŒ€í•´ platform ê¸°ì¤€ìœ¼ë¡œ ìš°ì„ ìˆœìœ„ë¥¼ ì ìš©í•˜ì—¬ ë‚¨ê¸¸ í–‰ì„ ê²°ì •
df_sorted = df.sort_values(by=['title', 'platform'], ascending=[True, True])
df_sorted['platform'] = pd.Categorical(df_sorted['platform'], categories=platform_order, ordered=True)

# ì¤‘ë³µëœ titleì„ ê°€ì§„ í–‰ì—ì„œ platform ìˆœì„œëŒ€ë¡œ ì¤‘ë³µ ì œê±°
df_deduplicated = df_sorted.drop_duplicates(subset=['title'], keep='first')

# ê²°ê³¼ í™•ì¸
print(f'ì¤‘ë³µ ì œê±° í›„ ë°ì´í„° í¬ê¸°: {df_deduplicated.shape}')
```

# 7. ì¥ë¥´ íŒŒì•… (listed_in)

'listed_in'ì—ì„œ ì¥ë¥´ë¥¼ ë¶„ë¦¬í•˜ê³  ì¶œí˜„ ë¹ˆë„ ê³„ì‚° ì™„ë£Œ, ìƒìœ„ 20ê°œ ì¥ë¥´ ì¶œë ¥. âœ…

```python
# 1ë‹¨ê³„: listed_in ì»¬ëŸ¼ì—ì„œ ì¥ë¥´ ëª©ë¡ì„ ì¶”ì¶œí•˜ì—¬ ì¥ë¥´ì˜ ì¶œí˜„ ë¹ˆë„ ê³„ì‚°í•˜ê¸°
# ê° ì˜í™”/TV í”„ë¡œê·¸ë¨ì˜ listed_inì´ ì—¬ëŸ¬ ê°œì˜ ì¥ë¥´ë¥¼ í¬í•¨í•  ìˆ˜ ìˆìœ¼ë¯€ë¡œ, ë¨¼ì € ì´ë¥¼ ë¶„ë¦¬í•˜ê³  ë‚˜ì—´ëœ ì¥ë¥´ë“¤ì„ í™•ì¸

# 'listed_in' ì»¬ëŸ¼ì—ì„œ ì¥ë¥´ë¥¼ ë¶„ë¦¬í•˜ì—¬ ë‚˜ì—´
all_genres = df_deduplicated['listed_in'].str.split(',').explode().str.strip()

# ì¥ë¥´ë³„ ì¶œí˜„ ë¹ˆë„ ê³„ì‚°
genre_counts = all_genres.value_counts()

# ê²°ê³¼ í™•ì¸
print(genre_counts.head(20))  # ìƒìœ„ 20ê°œì˜ ì¥ë¥´ë¥¼ ì¶œë ¥
```

# 8. ì „ì²´ ì¥ë¥´ ê³ ìœ ê°’ íŒŒì•…

'listed_in'ì—ì„œ ê³ ìœ í•œ ì¥ë¥´ ëª©ë¡ ì¶”ì¶œ ì™„ë£Œ. âœ…

```python
# 1. listed_in ì»¬ëŸ¼ì— ë“±ì¥í•˜ëŠ” ê³ ìœ í•œ ê°’ë“¤ì„ í™•ì¸
unique_genres = df_deduplicated['listed_in'].unique()

# 2. ê³ ìœ í•œ ì¥ë¥´ ëª©ë¡ ì¶œë ¥
print(unique_genres)
```

# 9. ì¥ë¥´ ê°œìˆ˜ íŒŒì•… ë° ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬í™”

'listed_in'ì—ì„œ ì„¸ë¶€ ì¥ë¥´ë¥¼ ë¶„ë¦¬í•˜ì—¬ ìœ ë‹ˆí¬í•œ ì¥ë¥´ 42ê°œ ì¶”ì¶œ ì™„ë£Œ. âœ…

```python
import re

# ì‘ì€ë”°ì˜´í‘œ ì œê±° í›„, ','ë¡œ ì„¸ë¶€ ì¥ë¥´ ë¶„ë¦¬í•˜ê¸°
cleaned_genres = []

for genre in unique_genres:
    # ì‘ì€ë”°ì˜´í‘œ ì œê±°
    genre = genre.replace("'", "")
    # ','ë¡œ ì„¸ë¶€ ì¥ë¥´ ë¶„ë¦¬
    cleaned_genres.extend(genre.split(', '))

# ì¤‘ë³µ ì œê±° í›„ ìœ ë‹ˆí¬í•œ ì„¸ë¶€ ì¥ë¥´ ì„¸ê¸°
unique_cleaned_genres = set(cleaned_genres)
num_unique_genres = len(unique_cleaned_genres)


print(f"ìœ ë‹ˆí¬í•œ ì„¸ë¶€ ì¥ë¥´ì˜ ê°œìˆ˜: {num_unique_genres}")
print("ìœ ë‹ˆí¬í•œ ì„¸ë¶€ ì¥ë¥´ ëª©ë¡:")
for genre in unique_cleaned_genres:
    print(genre)
```

# 10. 10ê°œ ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬ì™€ í•´ë‹¹ ì„¸ë¶€ ì¥ë¥´ ë§¤í•‘

10ê°œ ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬ì™€ ì„¸ë¶€ ì¥ë¥´ ë§¤í•‘ ì™„ë£Œ. âœ…

```python
genre_mapping = {
    'Action': [
        'Action & Adventure', 'TV Action & Adventure'  # 2ê°œ
    ],
    'Crime': [
        'Crime TV Shows', 'TV Mysteries'  # 3ê°œ
    ],
    'Sci-Fi/Fantasy': [
        'Sci-Fi & Fantasy', 'TV Sci-Fi & Fantasy', 'Anime Features', 'Anime Series', 'TV Sci-Fi & Fantasy'  # 5ê°œ
    ],
    'Comedy': [
        'Stand-Up Comedy', 'TV Comedies', 'Comedies', 'Stand-Up Comedy & Talk Shows'  # 4ê°œ
    ],
    'Horror/Thriller': [
        'TV Horror', 'Horror Movies', 'TV Thrillers', 'Thrillers', 'TV Action & Adventure', 'Cult Movies'  # 6ê°œ
    ],
    'War': [
        'War Movies'  # 1ê°œ
    ],
    'Sports': [
        'Sports Movies', 'Reality TV'  # 2ê°œ
    ],
    'Music/Musical': [
        'Music & Musicals', 'Stand-Up Comedy & Talk Shows'  # 2ê°œ
    ],
    'Romance/Love': [
        'Romantic TV Shows', 'Romantic Movies'  # 2ê°œ
    ],
    'Classic': [
        'Documentaries', 'Docuseries', 'Science & Nature TV', 'Independent Movies'  # 3ê°œ
    ]
}

# ê° ëŒ€ë¶„ë¥˜ ì¥ë¥´ë³„ ì†Œë¶„ë¥˜ ê°œìˆ˜ í™•ì¸
total_subgenres = 0
for category, subgenres in genre_mapping.items():
    print(f"{category}: {len(subgenres)} ê°œ")

    total_subgenres += len(subgenres)

print(f"\nì „ì²´ ì†Œë¶„ë¥˜ ì¥ë¥´ ê°œìˆ˜: {total_subgenres} ê°œ")
```

# 11. ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬í™”

listed_in ì»¬ëŸ¼ì„ ê¸°ë°˜ìœ¼ë¡œ ì„¸ë¶€ ì¥ë¥´ë¥¼ ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬ë¡œ ë§¤í•‘í•˜ì—¬ ìƒˆë¡œìš´ category ì»¬ëŸ¼ì„ ì¶”ê°€ âœ…

```python
# 4ï¸âƒ£ ì„¸ë¶€ ì¥ë¥´ë¥¼ ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜
def map_genres_to_categories(listed_in, genre_mapping):
    # ì‘ì€ë”°ì˜´í‘œ ì œê±°
    listed_in = listed_in.replace("'", "")

    # ê²°ê³¼ë¡œ ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬ ì œëª©ë§Œ ë‹´ì„ ë¦¬ìŠ¤íŠ¸
    categories = []

    # ì„¸ë¶€ ì¥ë¥´ë“¤ì„ ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬ë¡œ ë§¤í•‘
    for category, genres in genre_mapping.items():
        for genre in genres:
            if genre in listed_in:
                categories.append(category)
                break  # í•˜ë‚˜ì˜ ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬ë§Œ ì¶”ê°€ (ì¤‘ë³µ ë°©ì§€)

    # ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬ë“¤ì´ ì—†ìœ¼ë©´ ë¹ˆ ë¬¸ìì—´ ë°˜í™˜
    return ', '.join(categories) if categories else ''

# 5ï¸âƒ£ 'listed_in' ì»¬ëŸ¼ì„ ëŒ€ë¶„ë¥˜ ì¹´í…Œê³ ë¦¬ë¡œ ë³€í™˜
df_deduplicated['category'] = df_deduplicated['listed_in'].apply(lambda x: map_genres_to_categories(x, genre_mapping))

# 6ï¸âƒ£ ê²°ê³¼ í™•ì¸
print(df_deduplicated[['title', 'platform', 'listed_in', 'category']])
```

```python
import pandas as pd
from collections import Counter
from IPython.display import display

# df_deduplicatedë¥¼ ê¸°ë°˜ìœ¼ë¡œ ì‘ì—…
# ê° ì‘í’ˆì˜ categoryë¥¼ ì‰¼í‘œë¡œ ë¶„í• í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
df_deduplicated["category_list"] = df_deduplicated["listed_in"].apply(lambda x: x.split(", "))

# ëª¨ë“  categoryë¥¼ ëª¨ì•„ì„œ ë¹ˆë„ìˆ˜ ê³„ì‚°
all_categories = [cat for sublist in df_deduplicated["category_list"] for cat in sublist]
category_counts = Counter(all_categories)

# ì¹´í…Œê³ ë¦¬ë¥¼ ë¹ˆë„ìˆ˜ ê¸°ì¤€ìœ¼ë¡œ ì •ë ¬
top_categories = sorted(category_counts.items(), key=lambda x: x[1], reverse=True)
print("ì¹´í…Œê³ ë¦¬ ë¹ˆë„ìˆ˜ ìˆœìœ„:", top_categories)

# ê° ì‘í’ˆì— ëŒ€í•´ ê°€ì¥ ìš°ì„ ìˆœìœ„ê°€ ë†’ì€ ëŒ€ë¶„ë¥˜ ë°°ì •
def get_top_category(categories):
    sorted_categories = sorted(categories, key=lambda x: category_counts[x], reverse=True)
    return sorted_categories[0] if sorted_categories else None

df_deduplicated["top_category"] = df_deduplicated["category_list"].apply(get_top_category)

# ê²°ê³¼ ì¶œë ¥
print(df_deduplicated[["title", "top_category"]])

display(df_deduplicated.head(20))
```
